```python
import numpy as np
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
def notes(note, duration):
  global all
  
  if(note == 0):
    tmp, t, theta= phasor_sine(1, 10000, duration, C, 0)
    all = np.concatenate((all, tmp*0))
  else:
    tmp, t, theta= phasor_sine(1, 10000, duration, note, 0)
    all = np.concatenate((all, tmp))
```

```python
def score_to_notes(score_list):
  for note, dur in score_list:
    notes(note, dur)
```

```python
#amp = 1 sr = 10000 ìž„

C = 523
Bb = 466
A = 440
G = 392
F = 349
E = 329

all = np.empty(0)

score_list = [[C, 2*3/8], [Bb, 2*1/8], [A, 2*1/4],[G, 2*1/4],
              [F, 2*1/4],[G, 2*1/4],[ A, 2*1/4],[F, 2*1/4],
              [G, 2*1/8], [A, 2*1/8],[Bb, 2*1/8],[G, 2*1/8],[A, 2*3/8],[G, 2*1/8],
              [F, 2*1/4],[E, 2*1/4],[F, 2*1/4]]

score_to_notes(score_list)
```
